```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('VF_ALL.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].dtypes
```

```python
df['Stage'].unique()
```

```python
df['dayspost'].unique()
```

```python
df.head()
```

```python
df['Stage'] = df['Stage'].replace(0, 'PRE')
df['Stage'] = df['Stage'].replace('3 DAY POST', 'POST')
df['Stage'] = df['Stage'].replace('3 DAY IPSI', 'POST')
df['Stage'] = df['Stage'].replace('7 DAY', 'POST')
df['Stage'] = df['Stage'].replace('DAY  13', 'POST')
df['Stage'] = df['Stage'].replace('21 DAY', 'POST')
df['Stage'] = df['Stage'].replace('DAY24', 'POST')
df['Stage'] = df['Stage'].replace('DAY7', 'POST')
df['Stage'] = df['Stage'].replace('DAY3', 'POST')
df['Stage'] = df['Stage'].replace('DAYTERM', 'POST')
```
