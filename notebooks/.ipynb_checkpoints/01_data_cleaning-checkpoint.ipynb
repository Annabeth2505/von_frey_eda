```python
df['SurgeryDate'] = df.groupby('ID')['Date'].transform('min')
df['DayNumber'] = (df['Date'] - df['SurgeryDate']).dt.days

# Finds the earliest date for each unique 'ID' and then assign that earliest date to every row belonging to that 'ID'
```

```python
df.head()
```

```python
type(df)
```

```python
subset1 = df[(df['ID'] == 'DW368') & (df['Trial'] == 'I') & ( df['Stage'] == 'PRE') & (df ['DayNumber']== 0)]
```

```python
summary = subset1.groupby('Force')['Response'].mean().reset_index()
```

```python
sns.lineplot(data=summary, x='Force', y='Response', marker='o')
plt.title("DW368 - Day 0: Withdrawal Probability vs Force (Ipsilateral)")
plt.ylabel("P(Withdrawal)")
plt.grid(True)
plt.show()
```

```python
subset2 = df[(df['ID'] == 'DW369') & (df['Trial'] == 'I') & ( df['Stage'] == 'PRE') & (df ['DayNumber']== 0) & (df['Inconsistent']==False)]
```

```python
summary = subset2.groupby('Force')['Response'].mean().reset_index()
```

```python
sns.lineplot(data=summary, x='Force', y='Response', marker='o')
plt.title("DW369 - Day 0: Withdrawal Probability vs Force (Ipsilateral)")
plt.ylabel("P(Withdrawal)")
plt.grid(True)
plt.show()
```
